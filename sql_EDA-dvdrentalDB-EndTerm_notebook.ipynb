### Q1 - Write SQL to show 
- top 10 actors full names in terms of their appearance in films, 
- num of films they have acted in,
- num of language their films have released in
- the first year their films was released
- the longest film they have acted in

```python
### >>TODO<<:
ISQL = f'''SELECT 
    CONCAT(a.first_name, ' ', a.last_name) AS actor_name,
    COUNT(*) AS numFilms,
    COUNT(DISTINCT f.language_id) AS numLanguages,
    MIN(f.release_year) AS earliestReleaseYear,
    MAX(f.length) AS maxFilmLength
FROM 
    actor AS a
JOIN 
    film_actor AS fa ON a.actor_id = fa.actor_id
JOIN 
    film AS f ON fa.film_id = f.film_id
GROUP BY 
    CONCAT(a.first_name, ' ', a.last_name)
ORDER BY 
    numFilms DESC
LIMIT 
    10;'''
df1 = pd.io.sql.read_sql(ISQL, con=sqlEngine)
df1
```

### Q2 - Write SQL to count the number of films by category

```python
### >>TODO<<:
IISQL = f'''SELECT 
    c.name AS category_name,
    COUNT(*) AS numFilms
FROM 
    category AS c
JOIN 
    film_category AS fc ON c.category_id = fc.category_id
GROUP BY 
    c.category_id, c.name
ORDER BY 
    numFilms DESC;'''
df2 = pd.io.sql.read_sql(IISQL, con=sqlEngine)
df2
```

### Q3 - Write SQL to count the number films with each category and rating

```python
### >>TODO<<:
IIISQL = f'''SELECT 
    c.name AS category_name,
    f.rating,
    COUNT(*) AS numFilms
FROM 
    film AS f
JOIN 
    film_category AS fc ON f.film_id = fc.film_id
JOIN 
    category AS c ON fc.category_id = c.category_id
GROUP BY 
    c.category_id, c.name, f.rating
ORDER BY 
    SUBSTRING(c.name, 1, 1) ASC;'''
df3 = pd.io.sql.read_sql(IIISQL, con=sqlEngine)
df3.head(20)
```

### Q4 - Write SQL to count the number of actors for all films having 'G' rating within 'Travel' category using <b>film_list</b> table

```python
### >>TODO<<:
IVSQL = f'''SELECT 
    f.film_id,
    f.title,
    c.name AS category,
    f.rating,
    COUNT(DISTINCT fa.actor_id) AS numActors
FROM 
    film AS f
JOIN 
    film_category AS fc ON f.film_id = fc.film_id
JOIN 
    category AS c ON fc.category_id = c.category_id
JOIN 
    film_actor AS fa ON f.film_id = fa.film_id
WHERE 
    f.rating = 'G' AND c.name = 'Travel'
GROUP BY 
    f.film_id, f.title, c.name, f.rating;'''
df4 = pd.io.sql.read_sql(IVSQL, con=sqlEngine)
df4
```

### Q7 - Write SQL to count the total inventory by store and number of distinct films in its inventory

```python
### >>TODO<<:
VIISQL = f'''SELECT 
    s.store_id,
    COUNT(DISTINCT i.inventory_id) AS totalInventory,
    COUNT(DISTINCT i.film_id) AS distinctFilms
FROM 
    store AS s
JOIN 
    inventory AS i ON s.store_id = i.store_id
GROUP BY 
    s.store_id;'''
df7 = pd.io.sql.read_sql(VIISQL, con=sqlEngine)
df7
```

### Q8 - Write SQL to identify the countries that have more than 25 customer

```python
### >>TODO<<:
VIIISQL = f'''SELECT 
    c.country_id, c.country, COUNT(DISTINCT cu.customer_id) AS numCustomers
FROM 
    country AS c
JOIN 
    city AS ci ON c.country_id = ci.country_id
JOIN 
    address AS addr ON ci.city_id = addr.city_id
JOIN 
    customer AS cu ON addr.address_id = cu.address_id
GROUP BY 
    c.country_id, c.country
HAVING 
    COUNT(DISTINCT cu.customer_id) > 25;'''
df8 = pd.io.sql.read_sql(VIIISQL, con=sqlEngine)
df8
```

### Q10 - Write SQL to identify top 3 categoies (genres) inn term of total rental sales

```python
### >>TODO<<:
XSQL = f'''SELECT 
    c.name AS category,
    SUM(p.amount) AS total_sales
FROM 
    category AS c
JOIN 
    film_category AS fc ON c.category_id = fc.category_id
JOIN 
    film AS f ON fc.film_id = f.film_id
JOIN 
    inventory AS i ON f.film_id = i.film_id
JOIN 
    rental AS r ON i.inventory_id = r.inventory_id
JOIN 
    payment AS p ON r.rental_id = p.rental_id
GROUP BY 
    c.category_id, c.name
ORDER BY 
    total_sales DESC
LIMIT 
    3;'''
df10 = pd.io.sql.read_sql(XSQL, con=sqlEngine)
df10
```
